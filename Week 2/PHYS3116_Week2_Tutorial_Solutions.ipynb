````python
df1 = pd.DataFrame({
    'id': ['1', '2', '3'],
    'height': ['1', '2', '3']
})
df2 = pd.DataFrame({
    'star_id': ['Star 1', 'Star 2', 'Star 4'],
    'width': ['1', '2', '4']
})
````

To merge these two dataframes we need to complete a few operations:
1. Rename the column to be the same name to allow for the join
2. Update `id` column to mean the same format `Star X`
3. Save the dataframe

To rename columns, we can use 
`df.rename(columns={'old_column_name': 'new_column_name'}, inplace=True)`

````python
# renaming the columns in df1 to be 'star_id`
df1.rename(columns={'id': 'star_id'}, inplace=True)
df1
````

Apply an operation using `apply` method `df['column'] = df['column'].apply(func))`

````python
# Add Star to the start of the id column for dataframe 1 to match dataframe 2
def add_star(s):
    return "Star " + s
df1['star_id'] = df1['star_id'].apply(add_star)
df1
````

Merge dataframes on common column using `pd.merge`
Example: `df = pd.merge(df1, df2, on='star_id')`

````python
# Merge the two dataframes on 'star_id'
df = pd.merge(df1, df2, on='star_id')
df
````

Save to csv file for later using `df.to_csv('file_name.csv, index=False)`

````python
# Save id
df.to_csv('dataframe.csv', index=False)
````

### What does the above mean?
The velocity dispersion of a system is directly proportional to the luminosity of the system. As the luminosity decreases so does the velocity dispersion. Since luminosity is directly proportional to the stellar mass, we can see that the velocity dispersion is an excellent proxy for mass. We can in fact use the Virial Theorem to infer the dynamical mass of a system based on its velocity dispersion.

#### Virial Theorem  
The Virial Theorem relates the kenetic and potential energy of a system in dynamical equilibrium as follows:

$\rm{KE}=-\frac{1}{2}U$

Where $\rm{KE}$ is the kinetic energy and $\rm{U}$ is the gravitational potential energy.

$KE=\frac{1}{2}mv^2\\
U=-\frac{GMm}{r}$

$\rm M$ is the mass of the globular cluster and $\rm m$ is the mass of an individual star, which for simplicity we assume is the same across every star, $\rm r$ is the radius of the globular cluster, $\rm G$ is Newton's consant of gravity and $\rm v$ is the mean velocity of the star.

$\rm v^2=\frac{GM}{r}$

As we only get to measure a projected velocity for the stars in the system we get a radial velocity with respect to the globular cluster. The average of this radial velocity is the central velocity dispersion of the globular cluster $\rm v_{1D}^2\simeq\sigma_v^2$.

Another reasonable and simplifying assumption is that the velocity components are on average isotropic, i.e. there is no preferred direction of motion for each of the globular cluster components.

$v_{3D}^2=3v_{1D}^2=3\sigma_v^2$

Substituting this back into our relationship for the average velocity of a particle in a virialised system we get the following:

$3\sigma_v^2=\frac{GM}{r}\\
\frac{3r\sigma_v^2}{G}=M$

Hence we can estimate the mass of a globular cluster from the central velocity dispersion of the system.
